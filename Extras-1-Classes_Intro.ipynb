## Classes

Classes are an easy way to group <code>functions()</code> and <code>variables</code> together into an object that we can use. Once stored inside a class, we can look at variables kinda like a set of **attributes**, and functions kinda like a set of **actions**. 

That sounds all fine and dandy doesn't it - but what does that actually look like? Well, lets think about a class to represent a dog!

If a friend was telling you about their beloved pet, the things (or **attributes**) they'd probably mention are:
- The dog's name üè∑
- How old their pooch is üéÇ
- What breed the dog is üêï‚Äçü¶∫
- If their dog is cute ü•∫

And if you asked them what their dog loved to do, some activities (or **actions**) they'd probably mention are:
- Bark üîâ
- Dig üí™
- Walkies üö∂‚Äç‚ôÇÔ∏è

So lets take these and make a class to represent my dog: He is a **Daschund** called **Beans**. He is only **2** years old and *of course* he **is cute**!‚ú®‚ú®


<center></center>

Just like how we used the <code>def</code> keyword to tell python we're writing a function, we use the <code>class</code> keyword here, a **colon** and then **indent** everything we want to be inside the class.
Once the indent ends, any further code will not belong to the class.

```python
# Here's what a class containing my own dog's attributes stored as variables would look like
class Dog:
    name = "Beans"
    age = 2
    breed = "Daschund"
    cute = True
```
Notice how we used a mixture of **string**, **integer** and **boolean** variables? There are no restrictions to how you build your classes!

Now let's add the actions! We'll add these as functions to our class:

````python
# Run this cell to create the class so we can use it later!

class Dog:
    # Variables
    name = "Beans"
    age = 2
    breed = "Daschund"
    cute = True
    
    # Functions
    def bark():
        print("Woof!")        # This line of code is indented again, and so will only execute if the function above is called!
    
    def dig():
        print("Beans made a mess!")
    
    def walkies():
        print("Beans looks happy and tired now")
````

We can access the variables and functions inside the class by using the <code>.</code> operator. <br>
If I wanted to see what **breed** <code>Dog</code> is and if it's **cute** then I would use the cell below to print the attributes:


````python
print(Dog.breed, Dog.cute)
````
